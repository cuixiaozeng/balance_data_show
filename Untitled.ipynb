```python
import serial.tools.list_ports
import ctypes
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import time
import seaborn as sns
%matplotlib inline

# set up matplotlib
sns.set()
plt.rcParams['figure.figsize'] = (21.0, 8.0)
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
 from IPython import display

def h2f(s):
    cp = ctypes.pointer(ctypes.c_long(s))
    fp = ctypes.cast(cp,ctypes.POINTER(ctypes.c_float))
    return fp.contents.value
    
def change(data):
    result=[]
    result.append(h2f((data[4]<<24)|(data[3]<<16)|(data[2]<<8)|(data[1])))
    result.append(h2f((data[8]<<24)|(data[7]<<16)|(data[6]<<8)|(data[5])))
    result.append(h2f((data[12]<<24)|(data[11]<<16)|(data[10]<<8)|(data[9])))
    return result

def plot_durations(y):
 plt.figure(3)
 plt.clf()
 plt.subplot(311)
 plt.plot(y[:,0])
 plt.subplot(312)
 plt.plot(y[:,1])
 plt.subplot(313)
 plt.plot(y[:,2])
 plt.pause(0.001) # pause a bit so that plots are updated
 #if is_ipython:
 display.clear_output(wait=True)
 display.display(plt.gcf())
```

```python
plist = list(serial.tools.list_ports.comports())

if len(plist) <= 0:
    print("没有发现端口!")
else:
    plist_0 = list(plist[0])
    serialName = plist_0[0]
    serialFd = serial.Serial(serialName,128000,bytesize=8,stopbits=1,timeout=0.05)
    print("可用端口名>>>", serialFd.name)

if serialFd.isOpen():
    print("open success")
```

```python
y = []
while(1):
    try:
        data=serialFd.read(126)
        #time.sleep(0.01)        
        result=change(data)
        y.append(np.array([result[0],result[1],result[2]]))
        plot_durations(np.array(y))
    except:
        print("监控完毕！")
        break
```

```python
serialFd.close()
```
